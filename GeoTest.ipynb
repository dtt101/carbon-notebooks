# Visualising output over time

A test notebook to explore ways of visualising data over time per group

```python
# Import data
import pandas as pd

df = pd.read_excel('la-geo/pcan_data.xlsx', sheet_name='2019', header=3)

df.head(3)
```

```python
# Bar chart of total emissions for the UK
import matplotlib.pyplot as plt
import seaborn as sns

ukwide = df.iloc[0, [7, 8, 9, 10, 11, 12]]
fig = sns.barplot(x=ukwide.index, y=ukwide.values)
fig.set_xticklabels(fig.get_xticklabels(), rotation=45, horizontalalignment='right')
fig.set(xlabel='Category', ylabel='Total (per capita tonnes)', title='UK Emissions 2019, by category')
plt.show()
```

```python
# Example sub-category breakdown for LA (Darlington
darlo = df[df['Local Authority Code'] == 'E06000005']
darlo_food = darlo.iloc[0, [13, 14,15,16,17,18,19,20]]
fig = sns.barplot(x=darlo_food.index, y=darlo_food.values)
fig.set_xticklabels(fig.get_xticklabels(), rotation=45, horizontalalignment='right')
fig.set(xlabel='Food Category', ylabel='Total (per capita tonnes)', title='Darlington Local Authority - Food Emissions 2019, by category')
plt.show()
```
